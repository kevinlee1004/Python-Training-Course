## One plot
Set up figure object and add additional layes (plotting, lines, markers, colors, titles, axis, labels) to the figure object.  
- plot of x and y

```python
## import packages
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import xport  ## write sas datasets to local drive
```

```python
## Read SAS datasets
with open('./data/raw/vs.xpt', 'rb') as f: 
    _xport_vs = xport.to_columns(f)
    df_vs = pd.DataFrame(_xport_vs) 
```

```python
df_vs2 = df_vs[df_vs['VSTESTCD']=='DIABP'][df_vs['VSPOS']=='SUPINE']
df_vs2[['USUBJID','VISITNUM','VSSTRESN']]
```

```python
df_vs2_sub1 = df_vs2[df_vs2['USUBJID']=='01-701-1015']
```

```python
plt.figure()
```

### Plotting by x and y axis

```python
### plotting for Vital Sign results via visitnum
plt.plot(df_vs2_sub1['VISITNUM'], df_vs2_sub1['VSSTRESN'])
```

### Add lines, colors and markers

```python
### adding lines, colors and marker
plt.plot(df_vs2_sub1['VISITNUM'], df_vs2_sub1['VSSTRESN'], linestyle='--', color='r', marker = 'o')
```

### Add title and axis labels

```python
### adding title and axis labels
plt.plot(df_vs2_sub1['VISITNUM'], df_vs2_sub1['VSSTRESN'], linestyle='--', color='r', marker = 'o')
plt.title('Vital Sign for subject')
plt.xlabel('Visit Num')
plt.ylabel('Blood Pressure (mmHg)')
```

### Add another plotting at the same figure

```python
df_vs2_sub2 = df_vs2[df_vs2['USUBJID']=='01-701-1028']
```

```python
### Add another plotting and legend
plt.plot(df_vs2_sub1['VISITNUM'], df_vs2_sub1['VSSTRESN'], linestyle='--', color='r', marker = 'o', label='01-701-1015')
plt.plot(df_vs2_sub2['VISITNUM'], df_vs2_sub2['VSSTRESN'], linestyle='--', color='b', marker = 'o', label = '01-701-1028') ## add another subject 
plt.title('Vital Sign for subject')
plt.xlabel('Visit Num')
plt.ylabel('Blood Pressure (mmHg)')
plt.legend(loc='best') ## add legend at the best location
```

### Set up figure object for plots

```python
fig = plt.figure()
```

### Set up subplot of the figure

```python
fig = plt.figure()
subplot1 = fig.add_subplot(2,2,1) # 2 by 2 sub plot is 4 axes. Among 4 axes, this will be the first 
subplot2 = fig.add_subplot(2,2,2)
subplot3 = fig.add_subplot(2,2,3)
```

```python
## Set up figure object and sub-plots and plotting on each sub-plot should be done in the same cell to dispaly all the graph together.
## Set up figure object
fig = plt.figure(figsize=(10,10)) ## change figure size

## Set up sub plot of the figure
subplot1 = fig.add_subplot(2,2,1)
subplot2 = fig.add_subplot(2,2,2)
subplot3 = fig.add_subplot(2,2,3)

## First Sub-plot
df_vs2_sub = df_vs2[df_vs2['USUBJID']=='01-701-1015']
subplot1.plot(df_vs2_sub['VISITNUM'], df_vs2_sub['VSSTRESN'], linestyle='--', color='r', marker = 'o')
subplot1.set_title('Vital Sign for subject of 01-701-1015')
subplot1.set_xlabel('Visit Num')
subplot1.set_ylabel('Blood Pressure (mmHg)')

## Second Sub-plot
df_vs2_sub = df_vs2[df_vs2['USUBJID']=='01-701-1028']
subplot2.plot(df_vs2_sub['VISITNUM'], df_vs2_sub['VSSTRESN'], linestyle='--', color='r', marker = 'o')
subplot2.set_title('Vital Sign for subject of 01-701-1028')
subplot2.set_xlabel('Visit Num')
subplot2.set_ylabel('Blood Pressure (mmHg)')

## Third sub-plot
df_vs2_sub = df_vs2[df_vs2['USUBJID']=='01-718-1355']
subplot3.plot(df_vs2_sub['VISITNUM'], df_vs2_sub['VSSTRESN'], linestyle='--', color='r', marker = 'o')
subplot3.set_title('Vital Sign for subject of 01-701-1028')
subplot3.set_xlabel('Visit Num')
subplot2.set_ylabel('Blood Pressure (mmHg)')
```
